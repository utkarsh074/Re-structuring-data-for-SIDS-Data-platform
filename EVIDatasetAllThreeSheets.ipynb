```python
import json
import pandas as pd
import numpy as np

#Add path of files
dfEvi = pd.read_excel('C:\\Users\\utkarsh.dixit\\Downloads\\Datasets\\evi\\2021-retrospective-review-official.xlsx', sheet_name = ['Criteria','GNI','HAI','EVI'])
isoCountry = open('C:\\Users\\utkarsh.dixit\\PycharmProjects\\PovertyVulnerabilityAnalysis\\isonewcountryjs.json',)
countryDictionary = json.load(isoCountry)
```

```python
#Sheet1: Criteria
indicatorCode = ['GNI','HAI', 'EVI']
uniqueYears = dfEvi['Criteria']['Year'].unique()
dfEvi['Criteria'] = dfEvi['Criteria'].drop(columns = ['CCode','Status','Country'])
for ele in uniqueYears:
    dfEvi['Criteria'][ele] = np.nan
dfEvi['Criteria'].insert(2, 'IndicatorCode', '')
isoGrouped = dfEvi['Criteria'].groupby('Iso', sort = False)

Final_df = pd.DataFrame()
z = 0
for groupKey in isoGrouped.groups.keys():
    temp = isoGrouped.get_group(groupKey).apply(lambda x:x)
    temp = temp.reset_index(drop=True)
    for (code,i) in enumerate(indicatorCode,0):
        values = isoGrouped.get_group(groupKey)[i].tolist()
        temp.loc[code,uniqueYears] = values
        temp.loc[code,['IndicatorCode']] = i
    temp = temp.drop(columns = ['Year','GNI','HAI','EVI'])
    temp = temp.drop(range(3,temp.shape[0]))
    z = z + 1
    if z == 1:
        Final_df = temp
    else:
        Final_df = Final_df.append(temp, ignore_index=True)
        
Final_df.head(5)
```

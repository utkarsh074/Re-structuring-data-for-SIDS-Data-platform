```python
import json
import pandas as pd
import numpy as np

#Add path of files
dfEgov = pd.read_csv('C:\\Users\\utkarsh.dixit\\Downloads\\Datasets\\egov\\EGOV_DATA_2020.csv')
isoCountry = open('C:\\Users\\utkarsh.dixit\\PycharmProjects\\PovertyVulnerabilityAnalysis\\isonewcountryjs.json',)
countryDictionary = json.load(isoCountry)
```

```python
#truncate trailing whitespace
dfEgov['Country Name'] = dfEgov['Country Name'].str.strip()
dfEgov['CountryISO3'] = dfEgov['Country Name'].map(countryDictionary)
indicatorCode = ['E-Government Rank','E-Government Index','E-Participation Index',
                 'Online Service Index','Human Capital Index','Telecommunication Infrastructure Index']
indicatorCodeDict = {"E-Government Rank": "EGR", "E-Government Index":"EGI", "E-Participation Index":"EPI", 
                     "Online Service Index":"OSI","Human Capital Index":"HCI","Telecommunication Infrastructure Index":"TII"}
uniqueYears = dfEgov['Survey Year'].unique()

df = pd.DataFrame(columns = ['CountryISO3', 'IndicatorCode','2020'])
for row in dfEgov.iterrows():
    for code in indicatorCode:
        rowDict = {}
        rowDict.update([("CountryISO3", row[1]['CountryISO3'])])
        rowDict.update([("IndicatorCode", indicatorCodeDict.get(code))])
        rowDict.update([("2020", row[1][code])])
        df = df.append(rowDict, ignore_index=True)
df.head(12)
```
